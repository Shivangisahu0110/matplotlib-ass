Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in a wide range of formats. It provides a flexible and comprehensive toolkit for creating various types of plots, charts, graphs, and other data visualizations. Matplotlib is widely used in data analysis, scientific research, and data visualization tasks to help users understand data and communicate insights effectively.

Key features and uses of Matplotlib:

Data Visualization: Matplotlib is primarily used for creating high-quality 2D and 3D plots and charts to visualize data. It is suitable for both simple and complex data visualization tasks

Publication-Quality Graphics: Matplotlib provides fine-grained control over the appearance and formatting of plots, allowing users to create publication-quality graphics for research papers, reports, and presentations.

Wide Range of Plot Types: The Pyplot module of Matplotlib offers a variety of plot types, including but not limited to:

Line plots
Scatter plots
Bar charts
Histograms
Pie charts
Box plots
Heatmaps
Contour plots
3D plots
And many more.

Five common plots that can be created using the Pyplot module of Matplotlib:
    

Line Plot: A line plot (or line chart) is used to represent data points connected by straight lines, often used to visualize trends over time or continuous data.

Scatter Plot: Scatter plots display individual data points as markers on a 2D plane, making it useful for visualizing relationships between two variables or identifying patterns and outliers.

Bar Chart: Bar charts represent data using rectangular bars of varying heights or lengths, suitable for comparing categorical data or showing the distribution of data.

Histogram: A histogram is used to visualize the distribution of a single variable by dividing it into bins and displaying the frequency or density of data points in each bin.

Pie Chart: Pie charts display data as slices of a circular pie, with each slice representing a portion of the whole. They are often used to show the composition of a dataset or the relative proportions of different categories.

These are just a few examples of the many types of plots that can be created using Matplotlib's Pyplot module. Matplotlib's versatility and customization options make it a powerful tool for data visualization in Python.


Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane. It is commonly used to visualize the relationship between two variables and is particularly useful for identifying patterns, correlations, or clusters in data. Each data point is represented as a dot, and the position of the dot on the plot corresponds to the values of the two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y  ,c = "yellow",marker= "*")
plt.title("data points")
plt.xlabel("X - Label")
plt.ylabel("Y - Label")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

subplots method provides a way to plot multiple plots on a single figure. Given the number of rows and columns , it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax .

The subplots() function in the Matplotlib acts as a utility wrapper.This function helps in creating common layouts of subplots 

```python

import numpy as np
#For line 1: 
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
#For line 2:
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])
#For line 3: 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
#For line 4: 
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python

# Subplot 1
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, subplot 1
plt.plot(x1, y1)
plt.title('Plot 1')

# Subplot 2
plt.subplot(2, 2, 2)  # 2 rows, 2 columns, subplot 2
plt.plot(x2, y2)
plt.title('Plot 2')

# Subplot 3
plt.subplot(2, 2, 3)  # 2 rows, 2 columns, subplot 3
plt.plot(x3, y3)
plt.title('Plot 3')

# Subplot 4
plt.subplot(2, 2, 4)  # 2 rows, 2 columns, subplot 4
plt.plot(x4, y4)
plt.title('Plot 4')

# Add spacing between subplots
plt.tight_layout()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

A bar plot, also known as a bar chart or bar graph, is a graphical representation of data using rectangular bars or columns. Each bar's length or height is proportional to the value it represents. Bar plots are used to display and compare categorical data, making it easy to visualize and understand the distribution or relationship between different categories.

Bar plots are commonly used for the following purposes:

Comparing Data: Bar plots are useful for comparing the values of different categories or groups. They provide a clear visual representation of how one or more variables vary across categories.

Showing Frequencies: Bar plots can display the frequency or count of data points in different categories, helping to identify patterns and trends.

Displaying Rankings: Bar plots can be used to rank categories based on a particular variable. Categories with taller bars have higher values.

Tracking Changes Over Time: Stacked bar plots or grouped bar plots can be used to track changes in data over time or across different conditions

```python
# Horizontal bar plot
plt.barh(company,profit)
plt.title("Companies Profit (Horizontal bar plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Vertical bar plot
plt.bar(company,profit)
plt.title("Companies Profit (Horizontal bar plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
boxes = [box1,box2]
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays key summary statistics of the dataset, including the median, quartiles, and potential outliers. Box plots are useful for visualizing the spread and central tendency of data, as well as identifying any skewness or the presence of outliers.

Box plots are used for the following purposes:

Identifying the central tendency and spread of data.
Detecting skewness or asymmetry in the data distribution.
Highlighting potential outliers in the dataset.

```python
plt.boxplot(boxes)
plt.grid(True)
plt.show()
```
